### OPEN API를 이용하여 검색로 관련 게시물 불러오기

```python
import os
import sys
import urllib.request
import json
client_id = "yTZBtQrjCY9kFmf48Itk"
client_secret = "Qj6MEhaHek"
# 검색어 입력받기
quote = input('검색어를 입력해주세요.:')
encText = urllib.parse.quote(quote)
display_num = input("검색 출력결과 갯수를 적어주세요.(최대100,숫자만 입력):")
url = "https://openapi.naver.com/v1/search/blog?query=" + encText +"&display="+display_num# json 결과
# url = "https://openapi.naver.com/v1/search/blog.xml?query=" + encText # xml 결과
request = urllib.request.Request(url)
request.add_header("X-Naver-Client-Id",client_id)
request.add_header("X-Naver-Client-Secret",client_secret)
response = urllib.request.urlopen(request)
rescode = response.getcode()

if(rescode == 200):
    response_body = response.read()
    #print(response_body.decode('utf-8'))
else:
    print("Error Code:" + rescode)

body = response_body.decode('utf-8')
body
```

```python
# 불필요한 따옴표 제거
body = body.replace('"','')
```

```python
import re
titles = re.findall('title: (.*?),\nlink',body) # \n은 엔터를 의미
print('<<제목 모음>>')
print(titles)
print('총 제목 수: ',len(titles),'개') # 제목 갯수 확인
```

```python
# 블로그 링크 추출
links = re.findall('link: (.*?),\ndescription',body)
print('\n<<링크 모음>>')
print(links)
print('총 링크 수: ',len(links),'개') #링크 갯수 확인
```

```python
# 링크 다듬기 (불필요한 부분 제거)
blog_links = []

for i in links:
    a = i.replace('\\','')
    b = a.replace('?Redirect=Log&logNo=','/')
    blog_links.append(b)
    
print(blog_links)
print('생성된 링크 갯수:',len(blog_links),'개')
```

```python
import time
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
driver = webdriver.Chrome(os.path.abspath('C:/Users/user/Python/Crawling/chromedriver'))
driver.implicitly_wait(3)
```

```python
contents = []
for i in blog_links:
    #블로그 하나씩 불러오기
    driver.get(i)
    time.sleep(1)
    #블로그 안 본문이 있는 iframe에 접근하기
    driver.switch_to.frame("mainFrame")
    #본문 내용 크롤링하기
    a = driver.find_element_by_css_selector('div.se-main-container').text
    contents.append(a)
    #print('본문: \n',a)
    
# 창 닫기
driver.quit() 
print('<<본문 크롤링이 완료되었습니다.>>')
contents
```
